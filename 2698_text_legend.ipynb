```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
output_notebook()
plot = figure()
plot.line(x=[1, 2, 3], y=[0, 1, 0], legend="Change", line_width=2)
plot.circle(2, 2, legend="Peaks", color='green', line_width=4, fill_color="yellow", size=18)
plot.text(1, 1.5, text=['hello'], text_color="blue", legend="hello")
plot.text(2, 1.5, text=['world'], text_color="red", legend="world")  # legend not show, is this a bug?
show(plot)
```
