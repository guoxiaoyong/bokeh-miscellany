```python
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
output_notebook()
```

```python
data = {
    'x_1': ['apple', 'pear', 'plum', 'cat', 'dog', ], 
    'x_2': ['fruit', 'fruit', 'fruit', 'animal', 'animal'], 
    'y': [2, 1, 3, 5, 6]
       }
```

```python
from bokeh.models import ColumnDataSource
source = ColumnDataSource(data)
```

```python
from bokeh.models import FactorRange, DataRange1d
x_range_1 = FactorRange(factors=data['x_1'])
x_range_2 = FactorRange(factors=data['x_2'])
y_range = DataRange1d(start=-1, end=7)
```

```python
p = figure(x_range=x_range_1, y_range=y_range)
p.circle(x='x_1', y='y', radius=0.1, source=source)
show(p)
```

```python
p = figure(x_range=x_range_1, y_range=y_range)
p.extra_x_ranges['x'] = x_range_2
p.circle(x='x_1', y='y', radius=0.1, source=source)
show(p)
```

```python
p = figure(x_range=x_range_1, y_range=y_range)
p.add_layout(x_range_2, 'below')
p.circle(x='x_1', y='y', radius=0.1, source=source)
show(p)
```

```python
from bokeh.io import show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.palettes import RdBu3
from bokeh.plotting import figure
output_notebook()

c1 = RdBu3[2] # red
c2 = RdBu3[0] # blue
source = ColumnDataSource(dict(
    x=[1, 2, 3, 4, 5, 6],
    y=[2, 1, 2, 1, 2, 1],
    color=[c1, c2, c1, c2, c1, c2],
    label=['hi', 'lo', 'hi', 'lo', 'hi', 'lo']
))

p = figure(x_range=(0, 7), y_range=(0, 3), plot_height=300, tools='save')

# Note legend field matches the column in `source`
p.line( x='x', y='y', color='red', legend='label', source=source)
show(p)
```
